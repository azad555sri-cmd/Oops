```python
class computer():        
    
    def __init__(self, cpu, ram, price): # cpu & ram as argument
        print('in init')
    
    def config(self):     
        print('i5', '8gb', 50000)
        print('i7', 16, 100000)

com1 = computer('i5', 8, 50000)
com2 = computer('i7', 16, 10000)

com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 

    def config(self):     
        print('config is', self.cpu, self.ram)

com1 = computer('i5', 8)
com2 = computer('i7', 16)


com1.config() 
com2.config()
```

```python
class computer():        
    
    def __init__(self,cpu,ram, price): # cpu & ram as argument
        self.cpu = cpu #hear self is the object & we need to refer to the object
        self.ram = ram 
        self.price = price 

    def config(self):     
        print('config is', self.cpu, self.ram, self.price )

com1 = computer('i5', 8, 50000)
com2 = computer('i7', 16, 70000)


com1.config() 
com2.config()
```
